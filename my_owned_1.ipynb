```python
import pandas as pd
import re
from collections import Counter
import jieba

# 读取 CSV 文件到 DataFrame
file_path = '/Users/yautakchan/Desktop/Project/hktvmall_data.csv'
df = pd.read_csv(file_path)

# 定义一个函数来清除标点符号和空格
def clean_text(text):
    if isinstance(text, str):
        text = re.sub(r'[^\w]', '', text)  # 移除标点符号和空格
    return text

# 对 DataFrame 中的每个字符串列进行处理
for column in df.select_dtypes(include=['object']):
    df[column] = df[column].apply(clean_text)

# 合并所有文本数据到一个字符串中
all_text = ''.join(df.select_dtypes(include=['object']).fillna('').values.flatten())

# 使用 jieba 分词
words = jieba.lcut(all_text)

# 定义一个函数来生成 n-grams
def generate_ngrams(words, n):
    ngrams = zip(*[words[i:] for i in range(n)])
    return ["".join(ngram) for ngram in ngrams]

# 生成 n-grams
bigrams = generate_ngrams(words, 2)
trigrams = generate_ngrams(words, 3)
fourgrams = generate_ngrams(words, 4)
fivegrams = generate_ngrams(words, 5)

# 统计 n-grams 的出现频率
bigram_counter = Counter(bigrams)
trigram_counter = Counter(trigrams)
fourgram_counter = Counter(fourgrams)
fivegram_counter = Counter(fivegrams)

# 找出出现次数最多的 n-grams
most_common_bigram = bigram_counter.most_common(1)[0]
most_common_trigram = trigram_counter.most_common(1)[0]
most_common_fourgram = fourgram_counter.most_common(1)[0]
most_common_fivegram = fivegram_counter.most_common(1)[0]

# 打印出现次数最多的 n-grams 及其出现次数
print(f"出现最多的二元词组是：'{most_common_bigram[0]}', 出现次数为：{most_common_bigram[1]}")
print(f"出现最多的三元词组是：'{most_common_trigram[0]}', 出现次数为：{most_common_trigram[1]}")
print(f"出现最多的四元词组是：'{most_common_fourgram[0]}', 出现次数为：{most_common_fourgram[1]}")
print(f"出现最多的五元词组是：'{most_common_fivegram[0]}', 出现次数为：{most_common_fivegram[1]}")
```
