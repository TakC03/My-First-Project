```python
import pandas as pd
import re

# 读取CSV文件
df = pd.read_csv('/Users/yautakchan/Desktop/Project/success/cleaned_test.csv')

# 定义函数来提取香味
def extract_notes(description):
    notes = {
        '前調': [],
        '中調': [],
        '後調': []
    }
    
    # 正则表达式来匹配香味描述
    patterns = {
        '前調': r'前調\s*[:：]\s*(.*?)\n',
        '中調': r'中調\s*[:：]\s*(.*?)\n',
        '後調': r'後調\s*[:：]\s*(.*?)\n'
    }
    
    for key, pattern in patterns.items():
        match = re.search(pattern, description, re.DOTALL)
        if match:
            notes[key] = [note.strip() for note in match.group(1).split('，')]
    
    return notes

# 应用函数提取香味
df['Notes'] = df['Combined_Desc'].apply(extract_notes)

# 分离香味到不同列
df['前調'] = df['Notes'].apply(lambda x: x['前調'])
df['中調'] = df['Notes'].apply(lambda x: x['中調'])
df['後調'] = df['Notes'].apply(lambda x: x['後調'])

# 创建新的DataFrame仅包含Title和香味分类列
result_df = df[['Title', '前調', '中調', '後調']]

# 打印结果DataFrame
display(result_df)

# 如果需要保存到一个新的CSV文件
result_df.to_csv('classified_notes123456.csv', index=False)
```
